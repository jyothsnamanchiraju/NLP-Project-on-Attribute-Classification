```python
import json
#Loading non-lemmatized data from 'cleanSynopsis_ageRange_nonlemm.json'
# Opening the file 
with open("/gdrive/My Drive/MSC_Thesis_Project/Data/cleanSynopsis_ageRange_nonlemm.json") as file:
    # Load its content and make a new dictionary
    data = json.load(file)
```

```python
import pandas as pd

documents_df=pd.DataFrame(data['dataset'],columns=['desc','desc_sentences'])
#concatenating all rows of lemmatized data into a series of words - 
allSentences= []
sentence = ""
for i in documents_df.index:
         sentence = ""
         for a in  documents_df['desc'][i]:
             sentence =  sentence + a +" " 
         documents_df['desc_sentences'][i] = sentence
```

```python
import gensim.models as g
import codecs

#parameters
model="/gdrive/My Drive/MSC_Thesis_Project/Data/doc2vec.bin" # Load APNews dictionary
#inference hyper-parameters
start_alpha=0.01
infer_epoch=1000

#load model
m = g.Doc2Vec.load(model)
```

```python
import numpy as np
docs_df=pd.DataFrame(data['dataset'],columns=['desc'])

x =[]
for i in documents_df.index: #look up each doc in model    
    x.append(m.infer_vector(docs_df['desc'][i], alpha=start_alpha, steps=infer_epoch))
    
X= np.array(x)
```

```python
# Saving the doc2vec APNews features - 
np.save('/gdrive/My Drive/MSC_Thesis_Project/Data/doc2vec_apnews_features', X)  #will be saved as .npy file
```
