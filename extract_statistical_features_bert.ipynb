```python
from sentence_transformers import SentenceTransformer
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens') # can also try 'nli-distilroberta-base-v2'
```

```python
import numpy as np
docs_df=pd.DataFrame(data['dataset'],columns=['desc'])
x = []

for i in documents_df.index:
  try:
    #print(i)
    #print(documents_df['desc'][i])
    #print(len(documents_df['desc'][i]))    
    x.append(np.mean(sbert_model.encode(documents_df['desc'][i]),axis=0))
    #print(np.shape(x))
    #break
  except:
    print(i)
    print(documents_df['desc'][i])
  
X= np.array(x) 

```
